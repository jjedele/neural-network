```python
import numpy as np
import matplotlib.pylab as plt
from sklearn import datasets
from network import *
```

# Create Data Set

```python
n_classes = 3
X, Y = datasets.make_blobs(n_samples=1000, n_features=2, centers=n_classes)
Y_onehot = np.eye(3)[Y]
```

```python
plt.scatter(*X.T, c=Y)
```

# Train Model

```python
model = NeuralNetworkModel()

model.hidden_layer(units=10, activation=relu)
model.output_layer(units=n_classes, type=softmax_crossentropy)

final_loss = model.fit(X, Y_onehot)
print('Final loss: {}'.format(final_loss))
```

```python
# visualize training
l = model._cost_history
plt.plot(range(len(l)), l)
plt.ylabel('Loss')
plt.xlabel('Iteration')
```

# Visualize Model

```python
# visualize prediction
Y_pred = model.predict(X)
Y_pred_onedim = np.argmax(Y_pred, axis=1)
plt.scatter(*X.T, c=Y_pred_onedim)
```

## Decision Boundary

```python
x = np.arange(np.min(X[:,0]), np.max(X[:,0]), 0.1)
y = np.arange(np.min(X[:,1]), np.max(X[:,1]), 0.1)

X = np.array([[xx, yy] for xx in x for yy in y])
Y_pred = model.predict(X)
Y_pred_onedim = np.argmax(Y_pred, axis=1)

plt.scatter(*X.T, c=Y_pred_onedim)
```
