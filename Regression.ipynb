```python
import numpy as np
import matplotlib.pylab as plt
from sklearn import datasets
from network import *
```

# Create Data Set

```python
X, Y = datasets.make_regression(n_samples=100, n_features=1, noise=2.0, random_state=42)
```

```python
plt.scatter(X, Y)
```

# Train Model

```python
model = NeuralNetworkModel()

model.hidden_layer(units=5, activation=relu)
model.output_layer(units=1, type=identity_rmse)

final_loss = model.fit(X, Y.reshape((-1, 1)))
print('Final loss: {}'.format(final_loss))
```

```python
# visualize training
l = model._cost_history
plt.plot(range(len(l)), l)
plt.ylabel('Loss')
plt.xlabel('Iteration')
```

# Visualize Model

```python
# visualize prediction
Y_pred = model.predict(X)
plt.scatter(X, Y_pred)
```
