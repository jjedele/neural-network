```python
import numpy as np
import matplotlib.pylab as plt
from sklearn import datasets
from network import *
```

# Create Data Set

```python
X, Y = datasets.make_circles(n_samples=1000, noise=0.1, factor=0.4, random_state=42)
```

```python
plt.scatter(*X.T, c=Y)
```

# Train Model

```python
model = NeuralNetworkModel()

model.hidden_layer(units=10, activation=relu)
model.output_layer(units=1, type=sigmoid_binarycrossentropy)

final_loss = model.fit(X, Y.reshape((-1, 1)))
print('Final loss: {}'.format(final_loss))
```

```python
# visualize training
l = model._cost_history
plt.plot(range(len(l)), l)
plt.ylabel('Loss')
plt.xlabel('Iteration')
```

# Visualize Model

```python
# visualize prediction
Y_pred = model.predict(X)
plt.scatter(*X.T, c=Y_pred.flatten())
```

## Decision Boundary

```python
x = np.arange(-1, 1, 0.05)
y = np.arange(-1, 1, 0.05)

X = np.array([[xx, yy] for xx in x for yy in y])
Y_pred = model.predict(X)

plt.scatter(*X.T, c=Y_pred.flatten())
```
